# Part 2: Real Dataset — Titanic

```python
import pandas as pd
```

```python
data = pd.read_csv("Titanic-Dataset.csv")
```

### Step 1: Exploration

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

### Step 2: Data Cleaning

#### Handle missing values

```python
data.isnull().sum()
```

```python
data["Age"] = data["Age"].fillna(data["Age"].median())
```

```python
data["Age"].median()
```

```python
data.head(15)
```

```python
data["Embarked"].mode()
```

```python
data["Embarked"] = data["Embarked"].fillna(data["Embarked"].mode()[0])
```

```python
data.isnull().sum()
```

```python
data.drop(columns= ["Cabin"], inplace = True)
```

```python
data
```

### Step 3: Data Analysis

#### Survival rate by gender

```python
gender_survival = data.groupby('Sex')['Survived'].mean()
```

```python
gender_survival
```

```python
class_survival = data.groupby('Pclass')['Survived'].mean()
```

```python
class_survival
```

```python
avg_age_class = data.groupby('Pclass')['Age'].mean()
```

```python
avg_age_class
```

```python
data['AgeGroup'] = pd.cut(data['Age'], bins=[0, 12, 18, 60, 100], labels=['Child', 'Teen', 'Adult', 'Senior'])
```

```python
data.head(20)
```

```python
age_group_survival = data.groupby('AgeGroup', observed = False)['Survived'].mean()
```

```python
age_group_survival
```

### Step 4: Filtering

```python
female_survivors = data[(data['Sex'] == 'female') & (data['Survived'] == 1)]
```

```python
female_survivors
```

```python
children_survivors = data[(data['Age'] < 12) & (data['Survived'] == 1)]
```

```python
children_survivors
```

```python
high_prob_1st = data[(data['Pclass'] == 1) & (data['Survived'] == 1)]
```

```python
high_prob_1st
```

## Step 5: Insights

```python
print(f"1. Most likely to survive: Females (Rate: {gender_survival['female']:.2%})")
```

```python
print(f"2. Class affect: Yes, 1st Class had the highest survival ({class_survival[1]:.2%}) vs 3rd Class ({class_survival[3]:.2%})")
```

```python
print(f"3. Children Priority: Yes, Children (<12) had a survival rate of {age_group_survival['Child']:.2%}")
```

```python
print("4. Best Combination: 1st Class Females had a near-certain survival rate.")
```
